```python
# 2️⃣ GPU 정보 확인
import torch

print(f"GPU 이름: {torch.cuda.get_device_name(0)}")
print(f"전체 메모리: {torch.cuda.get_device_properties(0).total_memory / 1e9:.1f}GB")
print(f"CUDA 사용 가능: {torch.cuda.is_available()}")
```

```python
# 3️⃣ GPU 메모리 현황 출력 함수
def print_gpu_memory(label=""):
    """현재 GPU 메모리 사용량을 GB 단위로 출력"""
    used = torch.cuda.memory_allocated() / 1e9
    total = torch.cuda.get_device_properties(0).total_memory / 1e9
    print(f"[{label}] 사용 중: {used:.2f}GB / 전체: {total:.1f}GB (여유: {total-used:.1f}GB)")

# 시작 시점 메모리 확인
print_gpu_memory("시작 시점")
```
